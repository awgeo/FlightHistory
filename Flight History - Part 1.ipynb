# CrewHistoricData

## Concatenate various spreadsheets

```python
dir = '.\\\historic crew data\\'

path = os.getcwd()
files = os.listdir(dir)
files
```

```python
files_xlsx = [f for f in files if f[-4:] == 'xlsx']
files_xlsx
```

```python
df_hcd = pd.DataFrame()
```

```python
for f in files_xlsx:
    data = pd.read_excel('{}{}'.format(dir, f), 'Sheet1')
    df_hcd = df_hcd.append(data, ignore_index=False)
```

```python
df_hcd.head()
```

```python
df_hcd.shape
```

```python
df_hcd.drop(['Payroll', 'Name', 'Base'], axis=1, inplace=True)
```

```python
# Convert data types

df_hcd['Date'] = pd.to_datetime(df_hcd['Date'], format='%d/%m/%Y')
df_hcd['Duty Start'] = pd.to_timedelta(df_hcd['Duty Start']+':00')
df_hcd['Duty End'] = pd.to_timedelta(df_hcd['Duty End']+':00')
df_hcd['Block'] = pd.to_timedelta(df_hcd['Block']+':00')

df_hcd.dtypes
```

```python
df_hcd = df_hcd[df_hcd['Duty Code'] == 'FLY']
```

```python
df_hcd.shape
```

```python
df_hcd["Start Date"] = df_hcd["Date"]
df_hcd["End Date"] = df_hcd["Date"]
df_hcd["Temp"] = 0
df_hcd['Block Total'] = df_hcd['Block']
```

```python
# Reset index inplace so we can using index slicing below
df_hcd.reset_index(drop=True, inplace=True)

df_hcd.head()
```

```python
# Calculate overlap by interating over DataFrame rows as (index, Series) pairs
# Zero overlap for first Core Photo in each well, and no negative overlaps (representing gaps)

for index, row in df_hcd[1:].iterrows():
    
    current_row = df_hcd.loc[index]
    previous_row = df_hcd.iloc[df_hcd.index.get_loc(index) - 1]
    
    if current_row['Duty Start'] == previous_row['Duty End']:
        df_hcd.loc[index, 'Block Total'] = current_row['Block Total'] + previous_row['Block Total']
        
    if current_row['Duty Start'] == previous_row['Duty End']:
        df_hcd.loc[index, 'Duty Start'] = previous_row['Duty Start']
    
    if current_row['Flight'] == previous_row['Flight']:
        df_hcd.loc[index, 'Start Date'] = current_row['Start Date'] + pd.DateOffset(-1)
        
    if current_row['Destination'] == previous_row['Destination']:
        df_hcd.loc[index-1, 'Temp'] = previous_row['Temp'] + 1
    
df_hcd.round(1).head()
```

```python
df_hcd = df_hcd[df_hcd['Temp'] == 0]
```

```python
df_hcd.drop(['Date', 'Temp', 'Block', 'BTRT', 'Positioning', 'Category'], axis=1, inplace=True)
```

```python
#list(df_hcd.columns.values)

df_hcd[['Roster Month/Year',
 'Rank',
 'Start Date',
 'Duty Start',
 'End Date',
 'Duty End',
 'Flight',
 'Origin',
 'Destination',
 'Aircraft Type',
 'Block Total']].head()
```

```python
# 
df_hcd_airports  = pd.merge(df_hcd, df_airports[['IATA', 'City', 'Country', 'Latitude', 'Longitude']], how='left', left_on='Origin', right_on='IATA')

#
df_hcd_airports  = pd.merge(df_hcd_airports, df_airports[['IATA', 'City', 'Country', 'Latitude', 'Longitude']], how='left', left_on='Destination', right_on='IATA', suffixes=('_Origin', '_Destination'))

df_hcd_airports.head()
```

```python
pd.options.mode.chained_assignment = None

df_hcd_airports['Start'] = df_hcd_airports['Start Date'] + df_hcd_airports['Duty Start']
df_hcd_airports['End'] = df_hcd_airports['End Date'] + df_hcd_airports['Duty End']
#df_hcd_airports['Diff'] = (df_hcd_airports['End'] - df_hcd_airports['Start']) # correct!
```
